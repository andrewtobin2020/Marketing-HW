# Machine Learning - Logistic Regression Approach

```python
# My first approach will be with a logistic regression model. The first thing I need to do is encode the data
from sklearn.preprocessing import LabelEncoder
train_data_encoded = train_data.apply(LabelEncoder().fit_transform)
train_data_encoded.head()
```

```python
# assigning y 
y = train_data_encoded["churn"]
```

```python
# Assigning X
X = train_data_encoded.drop("churn", axis=1)
X.head()
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
# Fit the training data
classifier.fit(X_train, y_train)
```

```python
# Printing the training and testing scores
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}") 
```
